```python
training_imgs = ["{}.jpg".format(x) for x in list(training_set.id)]
```

```python
training_lbl = training_set['articleType']
```

```python
training_set = pd.DataFrame( {'Images': training_imgs,'Label': training_lbl})
```

```python
filter=list(data_dir.glob('*.*'))
```

```python
filter = [str(x).split('/')[3] for x in filter]
```

```python
filter
```

```python
training_set = training_set[training_set["Images"].isin(filter)]
```

```python
len(training_set)
```

```python
tset = training_set.groupby("Label").filter(lambda x: len(x) >200)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_dataGen = ImageDataGenerator(validation_split=0.1)

train_generator = train_dataGen.flow_from_dataframe(
                                        dataframe = tset,
                                        directory="/content/train/",x_col="Images",
                                        y_col="Label",
                                        class_mode="categorical",
                                        batch_size=444,subset='training')

validation_generator = train_dataGen.flow_from_dataframe(
    dataframe = tset,
    directory="/content/train/",
    x_col="Images",y_col = 'Label' ,batch_size=444,# same directory as training data
    class_mode='categorical',
    subset='validation')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Flatten, Dense, Dropout,GlobalMaxPooling2D
from tensorflow.keras.optimizers import Adam
import os
import tensorflow as tf
import datetime, os
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
from 
```

```python
checkpoint_filepath = '/home/ec2-user/'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)
```

```python
cnn_model = Sequential()
cnn_model.add(Conv2D(32, (3, 3), input_shape = (300,400,3), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size = (2, 2)))
cnn_model.add(Dropout(0.25))

cnn_model.add(Conv2D(64, (3, 3), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size = (4 ,4)))
cnn_model.add(Dropout(0.25))

cnn_model.add(Conv2D(128, (3, 3), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size = (4, 4)))
cnn_model.add(Dropout(0.25))

cnn_model.add(Conv2D(256, (3, 3), activation='relu'))
cnn_model.add(GlobalMaxPooling2D())
cnn_model.add(Dropout(0.25))

cnn_model.add(Flatten())
cnn_model.add(Dense(units = 512, activation = 'relu'))
cnn_model.add(Dropout(0.25))
cnn_model.add(Dense(units = 40, activation = 'softmax')) 
```

```python
cnn_model.summary()
```

```python
cnn_model.compile(loss ='categorical_crossentropy', optimizer='adam' ,metrics =['accuracy'])
```

```python
cnn_model.fit_generator(train_generator,
    validation_data = validation_generator, 
    validation_steps = validation_generator.samples, epochs = 10, callbacks=[EarlyStopping(monitor='val_loss', patience = 3),model_checkpoint_callback])
```
